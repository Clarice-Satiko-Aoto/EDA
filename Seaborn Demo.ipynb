# Drawing graphs

## barplot

```python
res= sns.barplot(df["cyl"], df["mpg"])
plt.show()
```

```python
res = sns.barplot(x="cyl", y="mpg", data=df)
plt.show()
```

```python
res= sns.barplot(x='cyl', y='mpg', data=df, color='yellow')
plt.show()
```

```python
res= sns.barplot(x='cyl', y='mpg', data=df, palette="Set2")
plt.show()
```

```python
res= sns.barplot(x='gear', y='hp', data=df, palette="rocket")
plt.show()
```

```python
#Countplot

sns.countplot(x='cyl', data= df, palette='Set1')
```

```python
sns.countplot(x='carb', data=df, palette="Set2")
```

```python
sns.countplot(y='gear', data=df, palette='pastel')
```

```python
sns.countplot(x='gear', hue='cyl', data=df, palette='rocket')
```

# distplot

```python

sns.distplot(df.mpg, bins=10, color = 'g')
```

```python
sns.histplot(df.mpg, bins=10, color = 'g')
```

# heatmaps

```python
sns.heatmap(df.corr(), cbar=True, lw=0.5) #lw = linewidth
```

```python
#load Iris dataset from lib

iris = sns.load_dataset('iris')
iris.sample(5)
```

```python
iris["species"].value_counts()
```

```python
iris.describe(include="O")
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data= iris)
```

```python
#using hue function. it is seen look like groupby function

sns.scatterplot(x="sepal_length", y="petal_length", hue= "species", data= iris)
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue= "species",size='species', data= iris)
```

# pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette='Set1')
```

# lmplot

```python
sns.lmplot(x='petal_length', y='petal_width', data=iris)
```

```python
sns.lmplot(x='petal_length', y="petal_width", data=iris, hue="species", markers=['o', '*', '^'])
```

# boxplot

```python
sns.boxplot(x="species", y="sepal_width", data=iris)
```
