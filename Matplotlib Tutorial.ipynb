```python
#Draw countour image

import matplotlib
import numpy as np
import matplotlib.cm as cm #color maps
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y= np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2
```

```python
fig, ax=plt.subplots()
CS= ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Countour map')
```

```python
#Draw 3D surface image
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

#Specify the 3D graphics to draw, with projection='3d'
ax=fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4, cstride=4, lw=0)
```

# Practice example  - criate a pie chart

```python
import matplotlib.pyplot as plt

#Data to plot
languages="Python", "Java", "PHP", "JavaScrip", "C#", "C++"
populatity= [22.2,17.6,8.8,8,7.7,6.7]
colors= ['#1f77b4', "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
```

```python
# explode 1st slice
explode=(0.1, 0,0,0,0,0)
plt.pie(populatity)
```

```python
plt.pie(populatity, explode=explode, labels=languages, colors=colors,
       autopct="%1.1f", shadow=True, startangle=90)
```
